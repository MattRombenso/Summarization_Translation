```python
import requests
from bs4 import BeautifulSoup
from transformers import MarianMTModel, MarianTokenizer

# Load the pre-trained MarianMT model for translation
model = MarianMTModel.from_pretrained('Helsinki-NLP/opus-mt-fr-en')
tokenizer = MarianTokenizer.from_pretrained('Helsinki-NLP/opus-mt-fr-en')

def fetch_article_content(url):
    response = requests.get(url)
    soup = BeautifulSoup(response.content, 'html.parser')
    
    paragraphs = soup.find_all('p')
    article_text = ' '.join([para.get_text() for para in paragraphs])
    
    return article_text

def translate(text, max_length=512):
    inputs = tokenizer(text, return_tensors="pt", truncation=True)
    translation_ids = model.generate(
        inputs['input_ids'],
        max_length=max_length,
        num_beams=6,  
        length_penalty=1.5,  
        early_stopping=True
    )
    translation = tokenizer.decode(translation_ids[0], skip_special_tokens=True)
    
    return translation

def process_translation_url(url, translation_max_length=512):
    article_text = fetch_article_content(url)
    translation = translate(article_text, max_length=translation_max_length)
    
    return article_text, translation

# Example usage
news_url = "https://www.lesaffaires.com/opinions/les-crises-boursieres-et-financieres-sont-plus-frequentes/"  
article_text, translation = process_translation_url(news_url)

print("Original French Text:", article_text[:500], "...") 
print("\nTranslation to English:", translation)
```
